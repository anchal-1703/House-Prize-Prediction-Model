#### train-Test splitting how it works

```python
def split_train_test(data,test_ratio):
    np.random.seed(42)
    shuffled = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled[:test_set_size] 
    train_indices = shuffled[test_set_size:]
    return data.iloc[train_indices],data.iloc[test_indices]
```

```python
train_set,test_set=split_train_test(housing,0.2)
```

```python
print(len(train_set),len(test_set))
```

### builtin function

```python
from sklearn.model_selection import train_test_split
train_set,test_set= train_test_split(housing,test_size=0.2,random_state=42)
print(len(train_set),len(test_set))
```

```python
# shuffeled split so that during training set evry possiblity covers
from sklearn.model_selection import StratifiedShuffleSplit
shuff = StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train_index,test_index in shuff.split(housing,housing['CHAS']):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
```

```python
#difference between stratshuffeled and simple traintestsplit is that stratshuffeled split equal number of values in both test and train set
print(strat_test_set.head())
print(test_set.head())
```
